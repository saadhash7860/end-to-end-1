#### Choosing features(input) and target(label) from our dataset.

```python
x = df.loc[:,['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','sqft_above','sqft_basement','yr_built','yr_renovated']]
```

```python
x
```

```python
y = df.loc[:,['price']]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Creating a class for object LinearRegression 
model = LinearRegression()
```

```python
# Model Fitting:-
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
